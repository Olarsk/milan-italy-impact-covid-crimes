```python
# Drop irrelevant columns

df.drop(columns=['ITTER107','Flags','Flag Codes','TIPO_DATO_FOL'],inplace=True)
```

```python
# Start with territory column
df['Territory'].value_counts()
```

```python
# Filter data by national and regional 
national_data = df[df['Territory'] == 'Italy']
```

```python
# Analyze ETA1 column to see they have anything useful
df['ETA1'].value_counts()
```

```python
df['Age class'].value_counts()
```

```python
# Drop the ETA1 column
df.drop(columns=['ETA1'], inplace=True)
```

```python
# Analyze the time column
print(df['TIME'].value_counts())
```

```python
# Clean Age column
df['Age class'].value_counts()
```

```python
df[['Lower_Age', 'Upper_Age']] = df['Age class'].str.split('-', expand=True)
```

```python
df.head(5)
```

```python
# Convert the age ranges into int
df['Lower_Age'] = df['Lower_Age'].astype(int)
df['Upper_Age'] = df['Upper_Age'].apply(lambda x: x.replace(' years', '') if isinstance(x, str) else x).astype(int)
```

```python
# Drop Age class column and TIME column
df.drop(columns=['Select time','Data type','Age class'],inplace=True)
```

```python
df.head(5)
```

```python
# Rename the value column to to unemployment column
df.rename(columns={'Value':'Unemployement_Rate','TIME':'Time_Period','SEXISTAT1':'Gender_ID'}, inplace=True)
```

```python
df.head(5)
```

```python
# Seperate them into regional and national 
national_data = df[df['Territory'] == 'Italy']
```

```python
national_data.rename(columns={'Territory': 'Country', 'Unemployement_Rate': 'NationalUNEM_R'}, inplace=True)
```

```python
national_data.head(5)
```

```python
regional_data = df[~df['Territory'].str.contains('Italy')]
```

```python
regional_data.head(5)
```

```python
regional_data.tail(5)
```

```python
# Rename some columns
regional_data.rename(columns={'Territory':'Province', 'Unemployement_Rate': 'RegionalUNEM_R'}, inplace=True)
```

```python
regional_data['Province'].value_counts()
```

```python
regional_data
```

```python
# Combine the data frame
unemployment_byprovince = pd.merge(national_data, regional_data,on=['Gender','Time_Period','Lower_Age','Upper_Age','Time_Period','Gender_ID'], how='left')
```

```python
unemployment_byprovince
```

```python
unemployment_byprovince.to_csv('unemployment_byprovince.2011_2022.csv',index=False)
```
