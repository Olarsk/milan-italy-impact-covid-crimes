```python
# identify if it has outliers
covid_data.describe(include='all')
```

# Feature engineering:

```python
covid_data['WHO_region'].unique()
```

```python
# create a dictionary
region = {'AFRO': 'Africa', 'AMRO': 'Americas', 'EMRO': 'Eastern Mediterranean', 'EURO': 'Europe', 'SEARO': 'South-East Asia', 'WPRO': 'Western Pacific', 'Other': 'Other'}

# create a new column, Region
covid_data['Region'] = covid_data['WHO_region'].map(region)
```

```python
# create a code for Region
covid_data['Region_code'] = covid_data['Region'].astype('category').cat.codes
```

```python
# create a code for Country
covid_data['Country_code'] = covid_data['Country_code'].astype('category').cat.codes
```

```python
# create new columns to separate the date
covid_data['Year'] = pd.DatetimeIndex(covid_data['Date_reported']).year
covid_data['Month'] = pd.DatetimeIndex(covid_data['Date_reported']).month
covid_data['Day'] = pd.DatetimeIndex(covid_data['Date_reported']).day

# drop the column, Date_reported
#covid_data.drop('Date_reported', axis=1, inplace=True)

covid_data.info()
```

```python
# reorder the columns
covid_data = covid_data[['Year','Month','Day','Country_code','Country','Region_code','Region','WHO_region','New_cases','Cumulative_cases','New_deaths','Cumulative_deaths']]
covid_data
```

```python
# save to csv file
covid_data.to_csv('../../data/WHO-COVID-19-global-data-cleaned.csv', index=False)
```
