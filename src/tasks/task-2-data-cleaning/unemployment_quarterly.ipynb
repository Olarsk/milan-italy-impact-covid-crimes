for `Flag Codes` and `Flags` will be drop.</br>

```python
# Droping the columns that are not needed
unemployment_qr_clean = unemployment_qr
unemployment_qr_clean.drop(['Select time','ETA1','TIPO_DATO_FOL', 'Data type', 'Flag Codes', 'Flags'], axis=1, inplace=True)
```

```python
# import regex
import re
```

```python
def extract_ages(age_str):
    ages = age_str.split('-')
    age_from = int(ages[0])
    age_to = int(ages[1].split()[0])
    return age_from, age_to
```

```python
unemployment_qr_clean[['ages from', 'ages to']] = unemployment_qr_clean['Age class'].apply(lambda x: pd.Series(extract_ages(x)))
unemployment_qr_clean.drop('Age class', axis=1, inplace=True)
unemployment_qr_clean.head()
```

```python
def extract_year_quarter(year_str):
    if '-Q' in year_str:
        year, quarter = year_str.split('-Q')
        return int(year), int(quarter)
    else:
        return int(year_str), None


unemployment_qr_clean[['Year', 'Quarter']] = unemployment_qr_clean['TIME'].apply(lambda x: pd.Series(extract_year_quarter(x)))
unemployment_qr_clean.drop('TIME', axis=1, inplace=True)
unemployment_qr_clean
```

```python
# rename value to unemployment rate
unemployment_qr_clean.rename(columns={'Value': 'Unemployment rate', 'SEXISTAT1': 'Gender_ID', 'ITTER107': 'Territory_acr'}, inplace=True)
```

```python
unemployment_qr_clean.info()
```

```python
# Create int ID for Territory
unemployment_qr_clean['Territory_ID'] = unemployment_qr_clean['Territory'].astype('category').cat.codes
unemployment_qr_clean.head()
```

```python
# fix order of columns
unemployment_qr_clean = unemployment_qr_clean[['Territory_ID','Territory','Territory_acr','Gender_ID','Gender','ages from','ages to','Year','Quarter','Unemployment rate']]
```
