## Geo data

Import dependencies

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

Using `Geopandas`, read the `geojson` files and plot the the points.  
  
### Regions

```python
region = gpd.read_file('../../data/geo_data/regions.geojson')
region.plot()
```

```python
reg_df = gpd.GeoDataFrame(region)
reg_df
```

```python
reg_codes = reg_df[['reg_name', 'reg_istat_code_num']]
reg_codes
```

```python
reg_codes.to_csv('../../data/geo_data/region_codes.csv', index=False)
```

### Provinces

```python
prov = gpd.read_file('../../data/geo_data/provinces.geojson')
prov.plot()
```

```python
prov_df = gpd.GeoDataFrame(prov)
prov_df
```

```python
prov_codes = prov_df[['prov_name', 'prov_istat_code_num']]
prov_codes
```

```python
prov_codes.to_csv('../../data/geo_data/province_codes.csv', index=False)
```

### Municipalities

```python
munics = gpd.read_file('../../data/geo_data/municipalities.geojson')
munics.plot()
```

```python
munic_df = gpd.GeoDataFrame(munics)
munic_df
```

```python
munic_codes = munic_df[['name', 'op_id']]
munic_codes
```

```python
munics.to_csv('../../data/geo_data/municipality_codes.csv', index=False)
```
